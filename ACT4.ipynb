```python
import control as ctrl
import matplotlib.pyplot as plt

num = [1]
den = [1, 2, 2]
G = ctrl.tf(num, den)
print(G)
```

```python
poles = ctrl.poles(G)
print("Poles:", poles)
```

```python
zero=ctrl.zeros(G)
print("Zero:", zero)
```

```python
import matplotlib.pyplot as plt
ctrl.pzmap(G, grid=True)
plt.show()
```
