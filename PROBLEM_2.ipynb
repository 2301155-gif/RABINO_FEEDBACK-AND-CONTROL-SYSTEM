```python
import control
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Coefficients for the rational polynomial form of G(s)
num = [1]
den = [1, 2, 3, 2, 0]
```

```python
# Define the transfer function
G_s = control.TransferFunction(num, den)
```

```python
# Attempt to extract step response characteristics
try:
    info = control.step_info(G_s)
    rt = info.get('RiseTime', 'N/A')
    st = info.get('SettlingTime', 'N/A')
    os = info.get('Overshoot', 'N/A')
except:
    rt, st, os = 'Undefined', 'Undefined', 'Undefined'

print(f"System Analysis for G(s):")
print(f"Rise Time: {rt}")
print(f"Settling Time: {st}")
print(f"Peak Overshoot: {os}")
```

```python
# Generate a 2x2 visualization
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
```
