```python
import control
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the coefficients for G(s) = (s + 1) / (s^4 + 4s^3 + 7s^2 + 6s + 3)
num = [1, 1]
den = [1, 4, 7, 6, 3]
```

```python
# Create the system model using the control library
sys = control.TransferFunction(num, den)
```

```python
# Calculate transient response specifications
info = control.step_info(sys)
print(f"Rise Time: {info['RiseTime']:.4f} s")
print(f"Settling Time: {info['SettlingTime']:.4f} s")
print(f"Peak Overshoot: {info['Overshoot']:.2f} %")
```

```python

# Initialize a 2x2 figure for comprehensive system visualization
fig = plt.figure(figsize=(12, 10))
```

```python
# Plot 1: Step Response of the displacement X2
plt.subplot(2, 2, 1)
t_s, y_s = control.step_response(sys)
plt.plot(t_s, y_s)
plt.title('Plot 1: Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (X2)')
plt.grid(True)
```

```python
# Plot 3: Pole-Zero Map to evaluate stability
plt.subplot(2, 2, 3)
control.pzmap(sys, plot=True)
plt.title('Plot 3: Pole-Zero Map')
plt.grid(True)
```
