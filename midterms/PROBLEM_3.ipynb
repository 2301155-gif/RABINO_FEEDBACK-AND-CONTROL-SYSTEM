```python
import control
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the transfer function coefficients
num = [3]
den = [8, 12, 26, 18, 0]
G = control.TransferFunction(num, den)
```

```python
# Extract system characteristics using step_info
try:
    s_info = control.step_info(G)
    tr = s_info.get('RiseTime', 'N/A')
    ts = s_info.get('SettlingTime', 'N/A')
    os = s_info.get('Overshoot', 'N/A')
except:
    tr, ts, os = 'N/A', 'N/A', 'N/A'

print(f'Rise Time: {tr}')
print(f'Settling Time: {ts}')
print(f'Overshoot: {os}')
```

```python
# Visualization setup
t = np.linspace(0, 20, 1000)
plt.figure(figsize=(12, 10))
```

```python
# Plot 1: Step Response
plt.subplot(2, 2, 1)
t_s, y_s = control.step_response(G, t)
plt.plot(t_s, y_s)
plt.title('Step Response')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
```

```python
# Plot 3: Pole-Zero Map
plt.subplot(2, 2, 3)
control.pzmap(G, plot=True)
plt.title('Pole-Zero Map')
```
